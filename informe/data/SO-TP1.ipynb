```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ej = 'ej'
tarch = 'tarch'
tmax = 'tmax'
ns = 'ns'
```

```python
df = pd.read_csv('time-data.csv')
df = df.groupby([tarch, tmax, ej]).min().reset_index()
```

```python
plt.clf()
dfPrueba = df[((df.tarch == 1) | (df.tarch == 3) | (df.tarch == 5))
              & ((df.tmax == 1) | (df.tmax == 3) | (df.tmax == 5))]
g = sns.FacetGrid(data=dfPrueba, col=tarch, row=tmax, sharey=False, size=5)
g = g.map(sns.barplot, ej, ns)
plt.show()
```
